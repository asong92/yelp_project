### Turn review topic_id_50 and topic_prob_50 columns into features

```python


def turn_dist_feature(df):
    d1 = []
    for val in df['word_dist'].values:
        n = []
        for t in val:
            n.append(tuple(reversed(t)))
        d1.append(dict(n))
    return d1

d1 = turn_dist_feature(star_1)
d2 = turn_dist_feature(star_2)
d3 = turn_dist_feature(star_3)
d4 = turn_dist_feature(star_4)
d5 = turn_dist_feature(star_5)

d = d1+d2+d3+d4+d5
word_features = pd.DataFrame(d)
word_features.fillna(0, inplace=True)
word_features.head()
```

```python
star_1.reset_index(inplace=True, drop=True)
star_2.reset_index(inplace=True, drop=True)
star_3.reset_index(inplace=True, drop=True)
star_4.reset_index(inplace=True, drop=True)
star_5.reset_index(inplace=True, drop=True)

all_reviews = star_1.append([star_2, star_3, star_4, star_5])
all_reviews
```

```python
word_features.reset_index(inplace=True, drop=True)
all_reviews.reset_index(inplace=True, drop=True)
```
