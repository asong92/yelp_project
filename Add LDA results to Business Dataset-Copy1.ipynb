### Turn review topic_id_50 and topic_prob_50 columns into features

```python
# topic_id_50 tells us which of the 50 topics the review is most about and 
# topic_prob_50 tells us the probability the review belongs to the topic
topic_cols = []
for index, row in reviews.loc[:,'topic_id_50':'topic_prob_50'].iterrows():
    d={}
    d[row[0]]=row[1]
    topic_cols.append(d)

topic_features = pd.DataFrame(topic_cols)
topic_features.fillna(0, inplace=True)
topic_features.columns = map(lambda x : 'topic_' + str(int(x)), topic_features.columns.values)
topic_features.head()
```

```python
# add topic features to our review data
reviews = pd.concat([reviews, topic_features], axis=1)
reviews
```

```python
for col in reviews.columns:
    print col
    print reviews[col].dtype
```

```python
def change_to_int(df, col_names):
    for name in col_names:
        df[name] = df[name].astype(int)
    return

cols = ['stars', 'votes.cool', 'votes.funny', 'votes.useful', 'review_length']
change_to_int(reviews, cols)
reviews.dtypes
```

## Join the topic distribution to business data
#### Note: At this point, we have 3,358 restaurants and 10,000 reviews from which I will build my models.

```python
grouped_reviews = reviews.groupby('business_id').agg([np.sum, np.mean])
grouped_reviews.reset_index(inplace=True)
```

```python
grouped_reviews.head()
```

```python
grouped_reviews.columns
```

```python
review_means = grouped_reviews.xs('mean', level=1, axis=1)
review_means['business_id'] = grouped_reviews['business_id'].values
review_means
```

```python
review_means.columns = map(lambda x : 'avg_' + x, review_means.columns.values)
review_means
```

```python
restaurants
```

```python
restaurants = pd.merge(left=restaurants, right=review_means, how='inner', left_on='business_id', right_on='avg_business_id')
restaurants.to_pickle('restaurant/restaurant_with_topic_dist1.pkl')
restaurants
```
